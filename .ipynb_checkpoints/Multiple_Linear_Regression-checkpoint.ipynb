# Visualisasi Model 

```python
#feature
X=df[['total_bill', 'size']]
X=sm.add_constant(X)
#target 
y=df['tip']

#define model
model_ols=sm.OLS(y,X)

#fitting model
reg=model_ols.fit()

#show summary
print(reg.summary())
```

```python
from mpl_toolkits.mplot3d import Axes3D

intercept=0.6689
slope_1=0.0927
slope_2=0.1926

#scatterplot
fig=plt.figure(figsize=[10,8])
ax=fig.add_subplot(projection='3d')
ax.scatter(df['total_bill'],df['size'],df['tip'],color ='blue')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Size')
ax.set_zlabel('Tip')
plt.title('Linear Regression Plane',size=15)

#mesh grid for 3D plane
total_bill_range= np.linspace(df['total_bill'].min(),df['total_bill'].max(),100)
size_range= np.linspace(df['size'].min(),df['size'].max(),100)
total_bill_mesh, size_mesh=np.meshgrid(total_bill_range,size_range)

#calculate prediction by linear regression
tip_mesh = intercept + slope_1*total_bill_mesh +slope_2*size_mesh

#plot the plane
ax.plot_surface(total_bill_mesh, size_mesh, tip_mesh, color='red', alpha=0.5,label='Manual Linear Regression Plane')

plt.show()
```

# Residual Analysis and Multicolinearity

1. Check normality distribution (residual)
2. Check Gauss Markov Assumption
3. Check Multicolinearity

```python
#predict y
y_pred=reg.predict(X)
```

```python
df_result=pd.DataFrame({
    'y_actual':df['tip'],
    'y_pred' : y_pred,
    'residual' : df['tip']-y_pred
})
df_result.head()
```

# 1. Normality Assumption

```python
#mengecek normalitas dari residual
fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(12,6))
#histogram
sns.histplot(data=df_result, x='residual',kde=True,ax=ax[0])
ax[0].set_title('Histogram of Residuals')
#qqplot 
sm.qqplot(df_result['residual'],line='s', ax=ax[1])
ax[1].set_title('Q-Q Plot of Residuals')
plt.show()
```

```python
from scipy.stats import normaltest
```

```python
#cara lain mendapatkan residual
reg.resid
```

```python
stat, pvalue =normaltest(df_result['residual'])

if pvalue>0.05:
    print(f'pvalue={pvalue}. pvalue > 0.05. Reject H0. Dapat diasumsikan Residual terdistribusi normal.')
else:
    print(f'pvalue={pvalue}. pvalue <= 0.05. Reject H0. Dapat diasumsikan Residual tidak terdistribusi normal.')
```

- residual tidak terdistribusi normal, asumsi normalitas tidak terpenuhi

# 2. Gauss Markov Assumption
- rata-rata eror (residual) =0 (mendekati 0)
- Uncorellated Error (tidak terdapat korelasi antara hasil prediksi dengan residual)
- Homoscedasticity (constant variance of residual)

```python
#residual plot
sns.scatterplot(data=df_result, x='y_pred', y='residual')
plt.axhline(y=0,color='red',linestyle='--')
```

![](https://condor.depaul.edu/sjost/it223/documents/resid-plots.gif)

```python
#cek rata-rata residual
df_result['residual'].mean()
```

## Interpretasi

a. Rata-rata error = 6.603096939901751e-15 (mendekati 0) 
- terpenuhi, karena eror yang positif dan yang negatif bernilai sama

b. uncorrelated error
- terpenuhi, karena erornya tidak membentuk pola tertentu (dapat dilihat juga dari hasil Durbin Watson, nilainya 2.096 mendekati 2 sehingga tidak ada autokorelasi dari residual terhadap hasil prediksi)

c. Homoscedasticity
- tidak terpenuhi (Homoscedasticity), karena pada nilai prediksi tip yang rendah, variansi erornya cenderung kecil. Sedangkan pada nilai prediksi tip yang tinggi variansi erornya cenderung besar.

## VIF : Variance Inflation Factor 

$\text{VIF}=1/(1-R^2)$

- VIF > 4 --> terindikasi multicolinearity
- VIF >10 --> multicolinearity parah

$R^2$ tinggi maka nilai VIF tinggi<br>
$R^2$ rendah maka nilai VIF rendah


```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
X=X.drop(columns=['const'])
X.head()
```

```python
#menghitung VIF untuk total_bill (indeks kolom ke 0)
variance_inflation_factor(X.values,0)
```

```python
#menghitung VIF untuk smoker_No (indeks kolom ke 3)
variance_inflation_factor(X.values,3)
```

```python
#membuat list VIF dari semua kolom
[variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
```

```python
def calculate_vif(X):
    vif = pd.DataFrame({
        'feature': X.columns,
        'VIF' : [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
    }).sort_values(by='VIF',ascending=False)

    return vif
```

```python
calculate_vif(X)
```

```python
calculate_vif(X.drop(columns=['time_Dinner','size']))
```

Sekarang sudah tidak ada multicolinearity

1. Normal Assumption tidak dipenuhi (menurut uji normalitas)
2. Gauss Markov Assumption tidak dipenuhi(hanya rata-rata eror =0 dan uncorellated eror)
3. Multicolinearity tidak terpenuhi
